```python
# Webscraping
import bs4
from selenium import webdriver
```

```python
driver = webdriver.Chrome(executable_path='C:\\Webdriver\\chromedriver.exe')
driver.get('https://www.infomoney.com.br/')
element = driver.find_element_by_tag_name('body')
resposta = element.get_attribute('innerHTML')
site = bs4.BeautifulSoup(resposta, 'html.parser')
```

```python
total = site.find_all('a', class_ = 'title')
```

```python
for i in total:
    print(i.text.strip())
```

```python
import requests
import bs4
#driver = webdriver.Chrome()
#driver.get('https://exame.com/exame-in/podemos-e-vamos-crescer-mais-diz-fabio-hering-apos-nao-a-arezzo/')

site = 'https://exame.com/exame-in/podemos-e-vamos-crescer-mais-diz-fabio-hering-apos-nao-a-arezzo/'
dic = requests.get(site)
#element = dic.text()
#resposta = element.get_attribute('innerHTML')
site = bs4.BeautifulSoup(dic.text, 'html.parser')
```

```python
print(site.prettify())
```

```python
titulo = site.find('h1', class_ = 'article-title')
```

```python
titulo
```

```python
print(titulo.text.strip())
```

```python
data = site.find_all('div', class_='article-date', limit=1)
```

```python
for i in data:
    print(i.text.strip())
```

```python
texto = site.find_all('p')
```

```python
for t in texto:
    print(t.text.strip())
```
